# Setting Up SQLite In Colab

```python
import sqlite3
import pandas as pd

# Creating our investment database in memory
conn = sqlite3.connect('investment_firm.db')
cursor = conn.cursor()

print("Database connected successfully.")
print("We are building: Naviel Capital Investment Database")
```

 #### Creating First Table: Companies

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS companies (
    company_id      INTEGER PRIMARY KEY,
    company_name    TEXT NOT NULL,
    sector          TEXT,
    country         TEXT,
    founded_year    INTEGER,
    stage           TEXT,    -- Early, Growth, Mature
    status          TEXT     -- Portfolio, Exited, Prospect
)
''')

conn.commit()
print("Companies table created.")
```

#### Creating Financials Table

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS financials (
    financial_id    INTEGER PRIMARY KEY,
    company_id      INTEGER,
    fiscal_year     INTEGER,
    revenue         REAL,        -- in millions USD
    ebitda          REAL,        -- in millions USD
    net_income      REAL,
    total_debt      REAL,
    cash            REAL,
    employees       INTEGER,
    FOREIGN KEY (company_id) REFERENCES companies(company_id)
)
''')

conn.commit()
print("Financials table created.")
```

####  Creating Investments Table

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS investments (
    investment_id       INTEGER PRIMARY KEY,
    company_id          INTEGER,
    investment_date     TEXT,
    investment_amount   REAL,    -- millions USD invested
    equity_stake        REAL,    -- percentage owned
    entry_valuation     REAL,    -- company valuation at entry
    current_valuation   REAL,    -- current estimated valuation
    lead_partner        TEXT,
    FOREIGN KEY (company_id) REFERENCES companies(company_id)
)
''')

conn.commit()
print("Investments table created.")
```

####  Creating SaaS Metrics Table

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS saas_metrics (
    metric_id       INTEGER PRIMARY KEY,
    company_id      INTEGER,
    fiscal_year     INTEGER,
    fiscal_quarter  INTEGER,
    arr             REAL,    -- Annual Recurring Revenue (millions)
    mrr             REAL,    -- Monthly Recurring Revenue (millions)
    churn_rate      REAL,    -- percentage of customers lost
    nrr             REAL,    -- Net Revenue Retention (%)
    cac             REAL,    -- Customer Acquisition Cost (USD)
    ltv             REAL,    -- Customer Lifetime Value (USD)
    customers       INTEGER,
    FOREIGN KEY (company_id) REFERENCES companies(company_id)
)
''')

conn.commit()
print("SaaS metrics table created.")
```

```python
# Q2: Which company grew revenue fastest in 2024?
query('''
    SELECT
        c.company_name,
        f.fiscal_year,
        f.revenue,
        f.ebitda,
        ROUND(f.ebitda / f.revenue * 100, 1) AS ebitda_margin_pct
    FROM financials f
    JOIN companies c ON f.company_id = c.company_id
    WHERE f.fiscal_year = 2024
    ORDER BY f.revenue DESC
''', "2024 REVENUE RANKING WITH EBITDA MARGINS")
```

# Aggregations, Grouping & Finance Questions

##### GROUP BY: Portfolio-Level Questions

```python
# Q4: Total capital deployed by sector
query('''
    SELECT
        c.sector,
        COUNT(i.investment_id)              AS num_investments,
        ROUND(SUM(i.investment_amount), 1)  AS total_invested_millions,
        ROUND(AVG(i.equity_stake), 1)       AS avg_stake_pct,
        ROUND(SUM(i.current_valuation), 1)  AS total_current_value
    FROM investments i
    JOIN companies c ON i.company_id = c.company_id
    GROUP BY c.sector
    ORDER BY total_invested_millions DESC
''', "CAPITAL DEPLOYMENT BY SECTOR")
```

```python
# Q5: Revenue growth by company across all years
query('''
    SELECT
        c.company_name,
        MIN(f.revenue)  AS revenue_2022,
        MAX(f.revenue)  AS revenue_2024,
        ROUND(
            (MAX(f.revenue) - MIN(f.revenue))
            / MIN(f.revenue) * 100, 1
        )               AS total_growth_pct
    FROM financials f
    JOIN companies c ON f.company_id = c.company_id
    GROUP BY c.company_name
    ORDER BY total_growth_pct DESC
''', "REVENUE GROWTH 2022-2024 BY COMPANY")
```

```python
# Q6: Which partner manages the most value?
query('''
    SELECT
        i.lead_partner,
        COUNT(i.investment_id)              AS companies_managed,
        ROUND(SUM(i.investment_amount), 1)  AS capital_deployed,
        ROUND(SUM(i.current_valuation), 1)  AS portfolio_value,
        ROUND(
            SUM(i.current_valuation) /
            SUM(i.investment_amount), 2
        )                                   AS value_multiple
    FROM investments i
    GROUP BY i.lead_partner
    ORDER BY portfolio_value DESC
''', "PARTNER PERFORMANCE - VALUE UNDER MANAGEMENT")
```

##### HAVING: Filtering After Grouping
**WHERE** filters rows before grouping. **HAVING** filters after. Critical distinction.

```python
# Q7: Which companies consistently grew revenue every year?
# Only show companies where ALL years show positive EBITDA
query('''
    SELECT
        c.company_name,
        COUNT(f.fiscal_year)        AS years_tracked,
        MIN(f.ebitda)               AS lowest_ebitda,
        MAX(f.ebitda)               AS highest_ebitda,
        ROUND(AVG(f.ebitda), 2)     AS avg_ebitda
    FROM financials f
    JOIN companies c ON f.company_id = c.company_id
    GROUP BY c.company_name
    HAVING MIN(f.ebitda) > 0
    ORDER BY avg_ebitda DESC
''', "CONSISTENTLY PROFITABLE COMPANIES (All years EBITDA > 0)")
```

```python
# Q8: Flag high-debt companies - PE due diligence risk screen
query('''
    SELECT
        c.company_name,
        f.fiscal_year,
        f.total_debt,
        f.ebitda,
        ROUND(f.total_debt / f.ebitda, 2)   AS debt_to_ebitda,
        CASE
            WHEN f.total_debt / f.ebitda > 5 THEN "HIGH RISK"
            WHEN f.total_debt / f.ebitda > 3 THEN "MONITOR"
            ELSE "HEALTHY"
        END                                  AS debt_risk_flag
    FROM financials f
    JOIN companies c ON f.company_id = c.company_id
    WHERE f.fiscal_year = 2024
    ORDER BY debt_to_ebitda DESC
''', "DEBT RISK SCREENING — 2024")
```

**Why this matters:** Debt-to-EBITDA above 4 - 5x is a red flag in PE. Lenders typically won't finance an acquisition if this ratio is too high.

###### CASE WHEN: Business Logic In SQL

```python
# Q9: Investment performance classification
query('''
    SELECT
        c.company_name,
        i.investment_amount,
        i.current_valuation,
        ROUND(i.current_valuation / i.investment_amount, 2) AS moic,
        CASE
            WHEN i.current_valuation / i.investment_amount >= 3.0
                THEN "STAR - Exceptional (3x+)"
            WHEN i.current_valuation / i.investment_amount >= 2.0
                THEN "STRONG - Good Return (2-3x)"
            WHEN i.current_valuation / i.investment_amount >= 1.5
                THEN "PERFORMING - On Track (1.5-2x)"
            WHEN i.current_valuation / i.investment_amount >= 1.0
                THEN "WATCH - Below Target (1-1.5x)"
            ELSE
                "UNDERPERFORMING - Capital at Risk (<1x)"
        END AS performance_band
    FROM investments i
    JOIN companies c ON i.company_id = c.company_id
    ORDER BY moic DESC
''', "INVESTMENT PERFORMANCE - MOIC CLASSIFICATION")
```

**MOIC** = **Multiple on Invested Capital**. It's the primary return metric in PE. A 3x MOIC means every $1

invested returned $3. Top PE funds target 3x+ on their best deals.

# Window Functions
most powerful tool in financial SQL - and the most commonly tested in interviews

**What they do:** Calculate values across rows without collapsing them into groups. You keep every row AND get aggregate context alongside it.

**GROUP BY asks:** "What is the average revenue across all companies?"

**Window function asks:** "What is the average revenue - AND show me each company's revenue next to that average on the same row?"

Window Function: Ranking

```python
# Q11: Rank companies by revenue within each sector
query('''
    SELECT
        c.company_name,
        c.sector,
        f.fiscal_year,
        f.revenue,
        RANK() OVER (
            PARTITION BY c.sector
            ORDER BY f.revenue DESC
        ) AS sector_revenue_rank,
        RANK() OVER (
            ORDER BY f.revenue DESC
        ) AS overall_revenue_rank
    FROM financials f
    JOIN companies c ON f.company_id = c.company_id
    WHERE f.fiscal_year = 2024
    ORDER BY c.sector, sector_revenue_rank
''', "REVENUE RANKING - WITHIN SECTOR AND OVERALL (2024)")
```

Breaking down the syntax:

- RANK() - assigns a rank number
- OVER() - tells SQL this is a window function
- PARTITION BY sector - restart ranking for each sector
- ORDER BY revenue DESC - rank by revenue, highest first

Running Totals: Tracking Cash Deployment Over Time

```python
# Q12: Cumulative capital deployed over time
query('''
    SELECT
        c.company_name,
        i.investment_date,
        i.investment_amount,
        SUM(i.investment_amount) OVER (
            ORDER BY i.investment_date
        ) AS cumulative_deployed,
        COUNT(i.investment_id) OVER (
            ORDER BY i.investment_date
        ) AS deals_completed
    FROM investments i
    JOIN companies c ON i.company_id = c.company_id
    ORDER BY i.investment_date
''', "CUMULATIVE CAPITAL DEPLOYMENT OVER TIME")
```

This is the running total pattern - used in every fund reporting system to show how capital has been deployed over a fund's life.

**LAG & LEAD:** Year-Over-Year Growth

This is the most used window function in finance. **LAG** looks backward. **LEAD** looks forward.

```python
# Q13: Year-over-year revenue growth per company
query('''
    SELECT
        c.company_name,
        f.fiscal_year,
        f.revenue,
        LAG(f.revenue) OVER (
            PARTITION BY f.company_id
            ORDER BY f.fiscal_year
        ) AS prior_year_revenue,
        ROUND(
            (f.revenue - LAG(f.revenue) OVER (
                PARTITION BY f.company_id
                ORDER BY f.fiscal_year
            )) /
            LAG(f.revenue) OVER (
                PARTITION BY f.company_id
                ORDER BY f.fiscal_year
            ) * 100, 1
        ) AS yoy_growth_pct
    FROM financials f
    JOIN companies c ON f.company_id = c.company_id
    ORDER BY c.company_name, f.fiscal_year
''', "YEAR-OVER-YEAR REVENUE GROWTH PER COMPANY")
```

What LAG does: For each row, it looks back one row (within the same company, ordered by year) and returns that value. The first year shows NULL because there's no prior year.

 Percentile Rankings: Where Does Each Company Stand?

```python
# Q14: EBITDA margin percentile across portfolio
query('''
    SELECT
        c.company_name,
        f.fiscal_year,
        ROUND(f.ebitda / f.revenue * 100, 1)    AS ebitda_margin_pct,
        ROUND(PERCENT_RANK() OVER (
            PARTITION BY f.fiscal_year
            ORDER BY f.ebitda / f.revenue
        ) * 100, 0)                              AS percentile_rank,
        NTILE(4) OVER (
            PARTITION BY f.fiscal_year
            ORDER BY f.ebitda / f.revenue DESC
        )                                        AS quartile
    FROM financials f
    JOIN companies c ON f.company_id = c.company_id
    WHERE f.fiscal_year = 2024
    ORDER BY ebitda_margin_pct DESC
''', "EBITDA MARGIN PERCENTILE RANKING (2024)")
```

NTILE(4) divides companies into quartiles - Q1 is top 25%, Q4 is bottom 25%. PE firms use this constantly to benchmark portfolio companies against each other.

Moving Averages: SaaS Metrics Trend Analysis

This connects SQL directly back to my earlier Python work - but now done in a database.

```python
# Q15: Quarterly ARR with 4-quarter moving average
query('''
    SELECT
        c.company_name,
        s.fiscal_year,
        s.fiscal_quarter,
        s.arr,
        s.churn_rate,
        ROUND(AVG(s.arr) OVER (
            PARTITION BY s.company_id
            ORDER BY s.fiscal_year, s.fiscal_quarter
            ROWS BETWEEN 3 PRECEDING AND CURRENT ROW
        ), 2) AS arr_4q_moving_avg,
        ROUND(AVG(s.churn_rate) OVER (
            PARTITION BY s.company_id
            ORDER BY s.fiscal_year, s.fiscal_quarter
            ROWS BETWEEN 3 PRECEDING AND CURRENT ROW
        ), 2) AS churn_4q_moving_avg
    FROM saas_metrics s
    JOIN companies c ON s.company_id = c.company_id
    ORDER BY c.company_name, s.fiscal_year, s.fiscal_quarter
''', "QUARTERLY ARR & CHURN WITH 4-QUARTER MOVING AVERAGES")
```

ROWS BETWEEN 3 PRECEDING AND CURRENT ROW - this defines the window frame. It says: look at this row and the 3 rows before it. That gives you a rolling 4-quarter average - identical to the rolling means I built in Python, but done entirely in SQL.

- In your own words - what is the difference between GROUP BY and a window function?

The core difference between GROUP BY and window functions lies in how they handle rows and aggregation:

GROUP BY:

Collapses rows: It aggregates rows into summary statistics, meaning you get one result row for each group defined. For example, if you GROUP BY sector, you'll get one row per sector with aggregated values.
Asks questions like: "What is the average revenue across all companies?"
Window Functions:

Retains original rows: They calculate values across a set of table rows that are related to the current row, but without collapsing those rows into a single output row. You keep every original row and get the aggregate context alongside it.
Asks questions like: "What is the average revenue – AND show me each company's revenue next to that average on the same row?"
They are powerful because they allow you to perform calculations (like ranking, moving averages, or year-over-year growth) over a 'window' of rows while still returning all the original detail rows.

CTE: Clean Revenue Growth

```python
# Rewrite Q13 using CTEs — same result, far more readable
query('''
    WITH yearly_revenue AS (
        SELECT
            f.company_id,
            c.company_name,
            f.fiscal_year,
            f.revenue,
            f.ebitda
        FROM financials f
        JOIN companies c ON f.company_id = c.company_id
    ),
    growth_calc AS (
        SELECT
            company_name,
            fiscal_year,
            revenue,
            ebitda,
            LAG(revenue) OVER (
                PARTITION BY company_id
                ORDER BY fiscal_year
            ) AS prior_revenue,
            ROUND(
                (revenue - LAG(revenue) OVER (
                    PARTITION BY company_id
                    ORDER BY fiscal_year
                )) / LAG(revenue) OVER (
                    PARTITION BY company_id
                    ORDER BY fiscal_year
                ) * 100, 1
            ) AS yoy_growth_pct
        FROM yearly_revenue
    )
    SELECT
        company_name,
        fiscal_year,
        revenue,
        yoy_growth_pct,
        CASE
            WHEN yoy_growth_pct >= 50 THEN "HYPERGROWTH"
            WHEN yoy_growth_pct >= 25 THEN "STRONG GROWTH"
            WHEN yoy_growth_pct >= 10 THEN "STEADY"
            WHEN yoy_growth_pct IS NULL THEN "BASE YEAR"
            ELSE                           "SLOW"
        END AS growth_classification
    FROM growth_calc
    ORDER BY company_name, fiscal_year
''', "REVENUE GROWTH CLASSIFICATION USING CTEs")
```

much easier read. Each CTE has a clear purpose. A colleague can understand the logic without you explaining it.

Multi-Step CTE: PE Deal Screening

```python
# Full deal screening pipeline using chained CTEs
query('''
    WITH base_financials AS (
        -- Step 1: Get latest year financials only
        SELECT
            f.company_id,
            f.revenue,
            f.ebitda,
            f.net_income,
            f.total_debt,
            f.cash,
            ROUND(f.ebitda / f.revenue * 100, 1)    AS ebitda_margin,
            ROUND(f.total_debt / f.ebitda, 2)        AS debt_to_ebitda,
            f.revenue - LAG(f.revenue) OVER (
                PARTITION BY f.company_id
                ORDER BY f.fiscal_year
            )                                        AS revenue_added
        FROM financials f
        WHERE f.fiscal_year = 2024
    ),
    company_scores AS (
        -- Step 2: Score each company on PE criteria
        SELECT
            b.company_id,
            b.revenue,
            b.ebitda,
            b.ebitda_margin,
            b.debt_to_ebitda,
            -- Score 1: Profitability (0-3)
            CASE
                WHEN b.ebitda_margin >= 30 THEN 3
                WHEN b.ebitda_margin >= 20 THEN 2
                WHEN b.ebitda_margin >= 10 THEN 1
                ELSE 0
            END AS profitability_score,
            -- Score 2: Debt health (0-3)
            CASE
                WHEN b.debt_to_ebitda <= 1 THEN 3
                WHEN b.debt_to_ebitda <= 2 THEN 2
                WHEN b.debt_to_ebitda <= 4 THEN 1
                ELSE 0
            END AS debt_score,
            -- Score 3: Scale (0-3)
            CASE
                WHEN b.revenue >= 20 THEN 3
                WHEN b.revenue >= 10 THEN 2
                WHEN b.revenue >=  5 THEN 1
                ELSE 0
            END AS scale_score
        FROM base_financials b
    ),
    final_screen AS (
        -- Step 3: Calculate total score and verdict
        SELECT
            s.company_id,
            s.revenue,
            s.ebitda,
            s.ebitda_margin,
            s.debt_to_ebitda,
            s.profitability_score,
            s.debt_score,
            s.scale_score,
            (s.profitability_score +
             s.debt_score +
             s.scale_score)              AS total_score,
            CASE
                WHEN (s.profitability_score +
                      s.debt_score +
                      s.scale_score) >= 7
                    THEN "PRIORITY TARGET"
                WHEN (s.profitability_score +
                      s.debt_score +
                      s.scale_score) >= 5
                    THEN "WATCH LIST"
                ELSE
                    "PASS"
            END                          AS acquisition_verdict
        FROM company_scores s
    )
    SELECT
        c.company_name,
        c.sector,
        c.country,
        f.revenue,
        f.ebitda_margin,
        f.debt_to_ebitda,
        f.profitability_score,
        f.debt_score,
        f.scale_score,
        f.total_score,
        f.acquisition_verdict
    FROM final_screen f
    JOIN companies c ON f.company_id = c.company_id
    WHERE c.status = "Portfolio"
    ORDER BY total_score DESC
''', "PE ACQUISITION SCREENING - MULTI-FACTOR SCORECARD")
```

CTE + Window Function Combined: The Gold Standard

```python
# Portfolio momentum report — combining everything
query('''
    WITH saas_trends AS (
        -- Calculate quarter-over-quarter ARR growth
        SELECT
            company_id,
            fiscal_year,
            fiscal_quarter,
            arr,
            churn_rate,
            nrr,
            ROUND(arr - LAG(arr) OVER (
                PARTITION BY company_id
                ORDER BY fiscal_year, fiscal_quarter
            ), 2) AS arr_qoq_growth,
            ROUND((arr - LAG(arr) OVER (
                PARTITION BY company_id
                ORDER BY fiscal_year, fiscal_quarter
            )) / LAG(arr) OVER (
                PARTITION BY company_id
                ORDER BY fiscal_year, fiscal_quarter
            ) * 100, 1) AS arr_qoq_growth_pct
        FROM saas_metrics
    ),
    latest_quarter AS (
        -- Get most recent quarter only
        SELECT *
        FROM saas_trends
        WHERE fiscal_year = 2024
          AND fiscal_quarter = 4
    ),
    momentum_score AS (
        -- Score each company's current momentum
        SELECT
            company_id,
            arr,
            churn_rate,
            nrr,
            arr_qoq_growth_pct,
            CASE
                WHEN arr_qoq_growth_pct >= 15
                 AND churn_rate <= 3.0
                 AND nrr >= 120
                    THEN "ACCELERATING"
                WHEN arr_qoq_growth_pct >= 8
                 AND churn_rate <= 5.0
                 AND nrr >= 100
                    THEN "HEALTHY"
                WHEN arr_qoq_growth_pct >= 0
                    THEN "STABLE"
                ELSE
                    "DECELERATING"
            END AS momentum_class
        FROM latest_quarter
    )
    SELECT
        c.company_name,
        c.sector,
        m.arr             AS current_arr,
        m.churn_rate,
        m.nrr,
        m.arr_qoq_growth_pct,
        m.momentum_class,
        RANK() OVER (
            ORDER BY m.arr_qoq_growth_pct DESC
        ) AS momentum_rank
    FROM momentum_score m
    JOIN companies c ON m.company_id = c.company_id
    ORDER BY momentum_rank
''', "SAAS PORTFOLIO MOMENTUM REPORT - Q4 2024")
```

#### Q1 v 1.1: Revenue Comparison 2023 vs 2024 for Portfolio Companies

```python
query('''
    SELECT
        c.company_name,
        MAX(CASE WHEN f.fiscal_year = 2023 THEN f.revenue ELSE NULL END) AS revenue_2023,
        MAX(CASE WHEN f.fiscal_year = 2024 THEN f.revenue ELSE NULL END) AS revenue_2024,
        ROUND(
            (MAX(CASE WHEN f.fiscal_year = 2024 THEN f.revenue ELSE NULL END) -
             MAX(CASE WHEN f.fiscal_year = 2023 THEN f.revenue ELSE NULL END)), 1
        ) AS absolute_change,
        ROUND(
            ((MAX(CASE WHEN f.fiscal_year = 2024 THEN f.revenue ELSE NULL END) -
              MAX(CASE WHEN f.fiscal_year = 2023 THEN f.revenue ELSE NULL END)) /
             MAX(CASE WHEN f.fiscal_year = 2023 THEN f.revenue ELSE NULL END)) * 100, 1
        ) AS percentage_change
    FROM financials f
    JOIN companies c ON f.company_id = c.company_id
    WHERE c.status = 'Portfolio'
    GROUP BY c.company_name
    HAVING revenue_2023 IS NOT NULL AND revenue_2024 IS NOT NULL
    ORDER BY percentage_change DESC
''', "PORTFOLIO REVENUE GROWTH 2023 vs 2024")
```

#### Q1 v 1.2 Senior version of Q1.1

```python
# CTE approach — same result, different architecture
query('''
    WITH pivoted AS (
        SELECT
            company_id,
            MAX(CASE WHEN fiscal_year = 2023
                THEN revenue END)   AS revenue_2023,
            MAX(CASE WHEN fiscal_year = 2024
                THEN revenue END)   AS revenue_2024
        FROM financials
        GROUP BY company_id
    )
    SELECT
        c.company_name,
        p.revenue_2023,
        p.revenue_2024,
        ROUND(p.revenue_2024 - p.revenue_2023, 1)
                                    AS absolute_change,
        ROUND((p.revenue_2024 - p.revenue_2023)
              / p.revenue_2023 * 100, 1)
                                    AS percentage_change
    FROM pivoted p
    JOIN companies c ON p.company_id = c.company_id
    WHERE c.status = "Portfolio"
      AND p.revenue_2023 IS NOT NULL
      AND p.revenue_2024 IS NOT NULL
    ORDER BY percentage_change DESC
''', "SENIOR VERSION - YoY Revenue Comparison")
```

```python
query('''
    WITH company_avg_revenue AS (
        SELECT
            company_id,
            ROUND(AVG(revenue), 1) AS avg_revenue_all_years
        FROM financials
        GROUP BY company_id
    )
    SELECT
        c.company_name,
        f.fiscal_year,
        f.revenue,
        car.avg_revenue_all_years,
        ROUND(f.revenue - car.avg_revenue_all_years, 1) AS revenue_deviation,
        ROUND(
            (f.revenue - car.avg_revenue_all_years) / car.avg_revenue_all_years * 100, 1
        ) AS percentage_deviation,
        CASE
            WHEN (f.revenue - car.avg_revenue_all_years) / car.avg_revenue_all_years * 100 < -20
                THEN 'ANOMALY'
            ELSE 'NORMAL'
        END AS anomaly_flag
    FROM financials f
    JOIN companies c ON f.company_id = c.company_id
    JOIN company_avg_revenue car ON f.company_id = car.company_id
    WHERE c.status = 'Portfolio'
    ORDER BY c.company_name, f.fiscal_year
''', "REVENUE ANALYSIS WITH DEVIATION AND ANOMALY FLAG")
```

#### Q3 v 1.1: High-Momentum SaaS Investments

```python
query('''
    WITH saas_with_lagged_data AS (
        SELECT
            company_id,
            fiscal_year,
            fiscal_quarter,
            arr,
            churn_rate,
            LAG(arr, 4) OVER (
                PARTITION BY company_id
                ORDER BY fiscal_year, fiscal_quarter
            ) AS arr_four_quarters_ago,
            LAG(churn_rate, 4) OVER (
                PARTITION BY company_id
                ORDER BY fiscal_year, fiscal_quarter
            ) AS churn_four_quarters_ago
        FROM saas_metrics
    ),
    latest_quarter_analysis AS (
        SELECT
            company_id,
            arr AS current_arr,
            churn_rate AS current_churn,
            arr_four_quarters_ago,
            churn_four_quarters_ago,
            ROUND(
                (arr - arr_four_quarters_ago) / arr_four_quarters_ago * 100, 1
            ) AS arr_growth_pct,
            (churn_rate < churn_four_quarters_ago) AS churn_improved_flag
        FROM saas_with_lagged_data
        WHERE fiscal_year = (SELECT MAX(fiscal_year) FROM saas_metrics)
          AND fiscal_quarter = (SELECT MAX(fiscal_quarter) FROM saas_metrics WHERE fiscal_year = (SELECT MAX(fiscal_year) FROM saas_metrics))
          AND arr_four_quarters_ago IS NOT NULL -- Ensure we have data from 4 quarters ago
    )
    SELECT
        c.company_name,
        lq.arr_four_quarters_ago || 'M vs ' || lq.current_arr || 'M' AS arr_then_vs_now,
        lq.arr_growth_pct,
        lq.churn_four_quarters_ago || '% vs ' || lq.current_churn || '%' AS churn_then_vs_now,
        CASE
            WHEN lq.arr_growth_pct >= 20 AND lq.churn_improved_flag THEN 'HIGH MOMENTUM'
            ELSE 'NORMAL MOMENTUM'
        END AS momentum_verdict
    FROM latest_quarter_analysis lq
    JOIN companies c ON lq.company_id = c.company_id
    ORDER BY lq.arr_growth_pct DESC
''', "HIGH-MOMENTUM SAAS INVESTMENTS (Latest Q)")
```

#### Q4: Risk-Adjusted Partner Performance

```python
query('''
    WITH financials_2024 AS (
        SELECT
            company_id,
            fiscal_year,
            revenue,
            ebitda,
            total_debt,
            ROUND(total_debt / ebitda, 2) AS debt_to_ebitda
        FROM financials
        WHERE fiscal_year = 2024
    ),
    risk_adjusted_deals AS (
        SELECT
            i.lead_partner,
            i.company_id,
            i.investment_amount,
            i.current_valuation,
            ROUND(i.current_valuation / i.investment_amount, 2) AS moic
        FROM investments i
        JOIN financials_2024 f ON i.company_id = f.company_id
        WHERE f.debt_to_ebitda < 3 -- Only count companies with debt-to-EBITDA below 3 in 2024
    )
    SELECT
        lead_partner,
        COUNT(DISTINCT company_id) AS companies_managed_risk_adjusted,
        ROUND(SUM(investment_amount), 1) AS total_deployed_risk_adjusted,
        ROUND(AVG(moic), 2) AS avg_moic_risk_adjusted,
        RANK() OVER (ORDER BY ROUND(AVG(moic), 2) DESC) AS partner_rank
    FROM risk_adjusted_deals
    GROUP BY lead_partner
    ORDER BY partner_rank;
''', "RISK-ADJUSTED PARTNER PORTFOLIO")
```
