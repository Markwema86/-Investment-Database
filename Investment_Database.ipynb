# Setting Up SQLite In Colab

```python
import sqlite3
import pandas as pd

# Creating our investment database in memory
conn = sqlite3.connect('investment_firm.db')
cursor = conn.cursor()

print("Database connected successfully.")
print("We are building: Naviel Capital Investment Database")
```

 #### Creating First Table: Companies

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS companies (
    company_id      INTEGER PRIMARY KEY,
    company_name    TEXT NOT NULL,
    sector          TEXT,
    country         TEXT,
    founded_year    INTEGER,
    stage           TEXT,    -- Early, Growth, Mature
    status          TEXT     -- Portfolio, Exited, Prospect
)
''')

conn.commit()
print("Companies table created.")
```

#### Creating Financials Table

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS financials (
    financial_id    INTEGER PRIMARY KEY,
    company_id      INTEGER,
    fiscal_year     INTEGER,
    revenue         REAL,        -- in millions USD
    ebitda          REAL,        -- in millions USD
    net_income      REAL,
    total_debt      REAL,
    cash            REAL,
    employees       INTEGER,
    FOREIGN KEY (company_id) REFERENCES companies(company_id)
)
''')

conn.commit()
print("Financials table created.")
```

####  Creating Investments Table

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS investments (
    investment_id       INTEGER PRIMARY KEY,
    company_id          INTEGER,
    investment_date     TEXT,
    investment_amount   REAL,    -- millions USD invested
    equity_stake        REAL,    -- percentage owned
    entry_valuation     REAL,    -- company valuation at entry
    current_valuation   REAL,    -- current estimated valuation
    lead_partner        TEXT,
    FOREIGN KEY (company_id) REFERENCES companies(company_id)
)
''')

conn.commit()
print("Investments table created.")
```

####  Creating SaaS Metrics Table

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS saas_metrics (
    metric_id       INTEGER PRIMARY KEY,
    company_id      INTEGER,
    fiscal_year     INTEGER,
    fiscal_quarter  INTEGER,
    arr             REAL,    -- Annual Recurring Revenue (millions)
    mrr             REAL,    -- Monthly Recurring Revenue (millions)
    churn_rate      REAL,    -- percentage of customers lost
    nrr             REAL,    -- Net Revenue Retention (%)
    cac             REAL,    -- Customer Acquisition Cost (USD)
    ltv             REAL,    -- Customer Lifetime Value (USD)
    customers       INTEGER,
    FOREIGN KEY (company_id) REFERENCES companies(company_id)
)
''')

conn.commit()
print("SaaS metrics table created.")
```

```python
# Q2: Which company grew revenue fastest in 2024?
query('''
    SELECT
        c.company_name,
        f.fiscal_year,
        f.revenue,
        f.ebitda,
        ROUND(f.ebitda / f.revenue * 100, 1) AS ebitda_margin_pct
    FROM financials f
    JOIN companies c ON f.company_id = c.company_id
    WHERE f.fiscal_year = 2024
    ORDER BY f.revenue DESC
''', "2024 REVENUE RANKING WITH EBITDA MARGINS")
```

# Aggregations, Grouping & Finance Questions

##### GROUP BY: Portfolio-Level Questions

```python
# Q4: Total capital deployed by sector
query('''
    SELECT
        c.sector,
        COUNT(i.investment_id)              AS num_investments,
        ROUND(SUM(i.investment_amount), 1)  AS total_invested_millions,
        ROUND(AVG(i.equity_stake), 1)       AS avg_stake_pct,
        ROUND(SUM(i.current_valuation), 1)  AS total_current_value
    FROM investments i
    JOIN companies c ON i.company_id = c.company_id
    GROUP BY c.sector
    ORDER BY total_invested_millions DESC
''', "CAPITAL DEPLOYMENT BY SECTOR")
```

```python
# Q5: Revenue growth by company across all years
query('''
    SELECT
        c.company_name,
        MIN(f.revenue)  AS revenue_2022,
        MAX(f.revenue)  AS revenue_2024,
        ROUND(
            (MAX(f.revenue) - MIN(f.revenue))
            / MIN(f.revenue) * 100, 1
        )               AS total_growth_pct
    FROM financials f
    JOIN companies c ON f.company_id = c.company_id
    GROUP BY c.company_name
    ORDER BY total_growth_pct DESC
''', "REVENUE GROWTH 2022-2024 BY COMPANY")
```

```python
# Q6: Which partner manages the most value?
query('''
    SELECT
        i.lead_partner,
        COUNT(i.investment_id)              AS companies_managed,
        ROUND(SUM(i.investment_amount), 1)  AS capital_deployed,
        ROUND(SUM(i.current_valuation), 1)  AS portfolio_value,
        ROUND(
            SUM(i.current_valuation) /
            SUM(i.investment_amount), 2
        )                                   AS value_multiple
    FROM investments i
    GROUP BY i.lead_partner
    ORDER BY portfolio_value DESC
''', "PARTNER PERFORMANCE - VALUE UNDER MANAGEMENT")
```

##### HAVING: Filtering After Grouping
**WHERE** filters rows before grouping. **HAVING** filters after. Critical distinction.

```python
# Q7: Which companies consistently grew revenue every year?
# Only show companies where ALL years show positive EBITDA
query('''
    SELECT
        c.company_name,
        COUNT(f.fiscal_year)        AS years_tracked,
        MIN(f.ebitda)               AS lowest_ebitda,
        MAX(f.ebitda)               AS highest_ebitda,
        ROUND(AVG(f.ebitda), 2)     AS avg_ebitda
    FROM financials f
    JOIN companies c ON f.company_id = c.company_id
    GROUP BY c.company_name
    HAVING MIN(f.ebitda) > 0
    ORDER BY avg_ebitda DESC
''', "CONSISTENTLY PROFITABLE COMPANIES (All years EBITDA > 0)")
```

```python
# Q8: Flag high-debt companies - PE due diligence risk screen
query('''
    SELECT
        c.company_name,
        f.fiscal_year,
        f.total_debt,
        f.ebitda,
        ROUND(f.total_debt / f.ebitda, 2)   AS debt_to_ebitda,
        CASE
            WHEN f.total_debt / f.ebitda > 5 THEN "HIGH RISK"
            WHEN f.total_debt / f.ebitda > 3 THEN "MONITOR"
            ELSE "HEALTHY"
        END                                  AS debt_risk_flag
    FROM financials f
    JOIN companies c ON f.company_id = c.company_id
    WHERE f.fiscal_year = 2024
    ORDER BY debt_to_ebitda DESC
''', "DEBT RISK SCREENING — 2024")
```

**Why this matters:** Debt-to-EBITDA above 4 - 5x is a red flag in PE. Lenders typically won't finance an acquisition if this ratio is too high.

###### CASE WHEN: Business Logic In SQL

```python
# Q9: Investment performance classification
query('''
    SELECT
        c.company_name,
        i.investment_amount,
        i.current_valuation,
        ROUND(i.current_valuation / i.investment_amount, 2) AS moic,
        CASE
            WHEN i.current_valuation / i.investment_amount >= 3.0
                THEN "STAR - Exceptional (3x+)"
            WHEN i.current_valuation / i.investment_amount >= 2.0
                THEN "STRONG - Good Return (2-3x)"
            WHEN i.current_valuation / i.investment_amount >= 1.5
                THEN "PERFORMING - On Track (1.5-2x)"
            WHEN i.current_valuation / i.investment_amount >= 1.0
                THEN "WATCH - Below Target (1-1.5x)"
            ELSE
                "UNDERPERFORMING - Capital at Risk (<1x)"
        END AS performance_band
    FROM investments i
    JOIN companies c ON i.company_id = c.company_id
    ORDER BY moic DESC
''', "INVESTMENT PERFORMANCE - MOIC CLASSIFICATION")
```

**MOIC** = **Multiple on Invested Capital**. It's the primary return metric in PE. A 3x MOIC means every $1

invested returned $3. Top PE funds target 3x+ on their best deals.

# Window Functions
most powerful tool in financial SQL - and the most commonly tested in interviews

**What they do:** Calculate values across rows without collapsing them into groups. You keep every row AND get aggregate context alongside it.

**GROUP BY asks:** "What is the average revenue across all companies?"

**Window function asks:** "What is the average revenue - AND show me each company's revenue next to that average on the same row?"

Window Function: Ranking

```python
# Q11: Rank companies by revenue within each sector
query('''
    SELECT
        c.company_name,
        c.sector,
        f.fiscal_year,
        f.revenue,
        RANK() OVER (
            PARTITION BY c.sector
            ORDER BY f.revenue DESC
        ) AS sector_revenue_rank,
        RANK() OVER (
            ORDER BY f.revenue DESC
        ) AS overall_revenue_rank
    FROM financials f
    JOIN companies c ON f.company_id = c.company_id
    WHERE f.fiscal_year = 2024
    ORDER BY c.sector, sector_revenue_rank
''', "REVENUE RANKING - WITHIN SECTOR AND OVERALL (2024)")
```

Breaking down the syntax:

- RANK() - assigns a rank number
- OVER() - tells SQL this is a window function
- PARTITION BY sector - restart ranking for each sector
- ORDER BY revenue DESC - rank by revenue, highest first

Running Totals: Tracking Cash Deployment Over Time

```python
# Q12: Cumulative capital deployed over time
query('''
    SELECT
        c.company_name,
        i.investment_date,
        i.investment_amount,
        SUM(i.investment_amount) OVER (
            ORDER BY i.investment_date
        ) AS cumulative_deployed,
        COUNT(i.investment_id) OVER (
            ORDER BY i.investment_date
        ) AS deals_completed
    FROM investments i
    JOIN companies c ON i.company_id = c.company_id
    ORDER BY i.investment_date
''', "CUMULATIVE CAPITAL DEPLOYMENT OVER TIME")
```

This is the running total pattern - used in every fund reporting system to show how capital has been deployed over a fund's life.

**LAG & LEAD:** Year-Over-Year Growth

This is the most used window function in finance. **LAG** looks backward. **LEAD** looks forward.

```python
# Q13: Year-over-year revenue growth per company
query('''
    SELECT
        c.company_name,
        f.fiscal_year,
        f.revenue,
        LAG(f.revenue) OVER (
            PARTITION BY f.company_id
            ORDER BY f.fiscal_year
        ) AS prior_year_revenue,
        ROUND(
            (f.revenue - LAG(f.revenue) OVER (
                PARTITION BY f.company_id
                ORDER BY f.fiscal_year
            )) /
            LAG(f.revenue) OVER (
                PARTITION BY f.company_id
                ORDER BY f.fiscal_year
            ) * 100, 1
        ) AS yoy_growth_pct
    FROM financials f
    JOIN companies c ON f.company_id = c.company_id
    ORDER BY c.company_name, f.fiscal_year
''', "YEAR-OVER-YEAR REVENUE GROWTH PER COMPANY")
```

What LAG does: For each row, it looks back one row (within the same company, ordered by year) and returns that value. The first year shows NULL because there's no prior year.

 Percentile Rankings: Where Does Each Company Stand?

```python
# Q14: EBITDA margin percentile across portfolio
query('''
    SELECT
        c.company_name,
        f.fiscal_year,
        ROUND(f.ebitda / f.revenue * 100, 1)    AS ebitda_margin_pct,
        ROUND(PERCENT_RANK() OVER (
            PARTITION BY f.fiscal_year
            ORDER BY f.ebitda / f.revenue
        ) * 100, 0)                              AS percentile_rank,
        NTILE(4) OVER (
            PARTITION BY f.fiscal_year
            ORDER BY f.ebitda / f.revenue DESC
        )                                        AS quartile
    FROM financials f
    JOIN companies c ON f.company_id = c.company_id
    WHERE f.fiscal_year = 2024
    ORDER BY ebitda_margin_pct DESC
''', "EBITDA MARGIN PERCENTILE RANKING (2024)")
```

NTILE(4) divides companies into quartiles - Q1 is top 25%, Q4 is bottom 25%. PE firms use this constantly to benchmark portfolio companies against each other.

Moving Averages: SaaS Metrics Trend Analysis

This connects SQL directly back to my earlier Python work - but now done in a database.

```python
# Q15: Quarterly ARR with 4-quarter moving average
query('''
    SELECT
        c.company_name,
        s.fiscal_year,
        s.fiscal_quarter,
        s.arr,
        s.churn_rate,
        ROUND(AVG(s.arr) OVER (
            PARTITION BY s.company_id
            ORDER BY s.fiscal_year, s.fiscal_quarter
            ROWS BETWEEN 3 PRECEDING AND CURRENT ROW
        ), 2) AS arr_4q_moving_avg,
        ROUND(AVG(s.churn_rate) OVER (
            PARTITION BY s.company_id
            ORDER BY s.fiscal_year, s.fiscal_quarter
            ROWS BETWEEN 3 PRECEDING AND CURRENT ROW
        ), 2) AS churn_4q_moving_avg
    FROM saas_metrics s
    JOIN companies c ON s.company_id = c.company_id
    ORDER BY c.company_name, s.fiscal_year, s.fiscal_quarter
''', "QUARTERLY ARR & CHURN WITH 4-QUARTER MOVING AVERAGES")
```

ROWS BETWEEN 3 PRECEDING AND CURRENT ROW - this defines the window frame. It says: look at this row and the 3 rows before it. That gives you a rolling 4-quarter average - identical to the rolling means I built in Python, but done entirely in SQL.

Full Window Function Dashboard

```python
# Q16: Complete SaaS health scorecard using window functions
query('''
    SELECT
        c.company_name,
        s.fiscal_year,
        s.fiscal_quarter,
        s.arr,
        s.churn_rate,
        s.nrr,
        ROUND(s.ltv / s.cac, 2)             AS ltv_cac_ratio,
        RANK() OVER (
            PARTITION BY s.fiscal_year, s.fiscal_quarter
            ORDER BY s.nrr DESC
        )                                    AS nrr_rank,
        ROUND(s.arr - LAG(s.arr) OVER (
            PARTITION BY s.company_id
            ORDER BY s.fiscal_year, s.fiscal_quarter
        ), 2)                                AS arr_added_vs_prior_q,
        CASE
            WHEN s.nrr >= 120 THEN "ELITE - Best in class"
            WHEN s.nrr >= 110 THEN "STRONG - Above average"
            WHEN s.nrr >= 100 THEN "HEALTHY - Retaining revenue"
            ELSE                   "AT RISK - Losing revenue"
        END                                  AS nrr_classification
    FROM saas_metrics s
    JOIN companies c ON s.company_id = c.company_id
    ORDER BY c.company_name, s.fiscal_year, s.fiscal_quarter
''', "SAAS HEALTH SCORECARD WITH WINDOW FUNCTIONS")
```

**LTV/CAC ratio** - if it costs $ 1,000

 to acquire a customer (CAC) and they generate $10,000 over their lifetime (LTV), the ratio is 10x. Top SaaS companies target 3x minimum. Below 1x means you're losing money on every customer.

**NRR (Net Revenue Retention)** above 100% means existing customers are spending MORE over time - the holy grail of SaaS. It means the company grows even without acquiring new customers.
